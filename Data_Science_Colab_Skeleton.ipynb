# **Ingest**

[NBA Player Data](https://www.kaggle.com/datasets/vivovinco/nba-player-stats/?select=2021-2022+NBA+Player+Stats+-+Regular.csv) from 2021-2022 Regular Season

```python
import pandas as pd
import numpy as np
```

```python
file = "/content/2021-2022 NBA Player Stats - Regular.csv"
nba = pd.read_csv(file, encoding='Windows-1252',sep=';')
nba
```

```python
nba = nba.drop(columns=["Rk"])
nba
```

# **EDA**

### Data-Based EDA

```python
import seaborn.objects as so

so.Plot(nba,  x="FGA", y="FG").add(so.Dot()
  ).label(x="Average Field Goal Attempt", y="Average Field Goal Made", title="Display of Field Goal Efficiency in NBA 2021-2022 Season")
```

```python
nba[(nba["FGA"] > 20) & (nba["FG"] > 10)].sort_values(by=["FG","FGA"], ascending=False)
```
